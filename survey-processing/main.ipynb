# AlgoQuest - Survey processing

```python
import pandas as pd

data = pd.read_csv('data/survey.csv')
data.info()

data.head()
```

### Merge generic algorithm columns with specific ones

```python
data.iloc[:, 14] = data.iloc[:, 14].fillna(data.iloc[:, 8])
data.iloc[:, 15] = data.iloc[:, 15].fillna(data.iloc[:, 9])
data.iloc[:, 16] = data.iloc[:, 16].fillna(data.iloc[:, 10])

data.iloc[:, 17] = data.iloc[:, 17].fillna(data.iloc[:, 11])
data.iloc[:, 18] = data.iloc[:, 18].fillna(data.iloc[:, 12])
data.iloc[:, 19] = data.iloc[:, 19].fillna(data.iloc[:, 13])

data.drop(data.columns[[8, 9, 10, 11, 12, 13]], axis=1, inplace=True)
```

### Calculate SUS score

```python
data['sus_score'] = 2.5 * ((data.iloc[:, 14:24:2].sum(axis=1) - 5) + (25 - data.iloc[:, 15:24:2].sum(axis=1)))

data.drop(data.columns[14:24], axis=1, inplace=True)

data.head()
```

### Save processed data

```python
data.drop(['Timestamp', 'Ποια είναι η εμπειρία σου με αλγόριθμους;', 'Ποια είναι η εμπειρία σου με παιχνίδια;'], axis=1, inplace=True)

data.to_csv('data/survey_processed.csv', index=False)
```
